# PyMongo | Breast Cancer Data | BDA Miniproject

### MongoDB Connection

```python
#1) import mongo client

from pymongo import MongoClient
import pandas as pd
```

```python
#2) create an instance of the client

client=MongoClient()
```

```python
#3) establish connection

Client= MongoClient('localhost',27017)
```

```python
#4) create database 

db = client.BDAmini
```

```python
#5) create collection

cancer=db.cancer
```

```python
#6) import the dataset

df=pd.read_csv("data.csv")
```

```python
#7) Convert to documents

records_=df.to_dict(orient='records')
```

```python
#8) insert records into collection

result=db.cancer.insert_many(records_)
```

```python
#9) show list of existing database

print(client.list_database_names())
```

```python
#10) show list of existing collections

print(client["BDAmini"].list_collection_names())
```

### Queries

```python
#1) print total number of patient records

cancer.count_documents({})
```

```python
#2) Display all records
from pprint import pprint

for e in cancer.find():
    pprint(e)
```

```python
#3) display the first 5 records with diagnosis "Malignant"

canquery = { 'diagnosis': 'M' }
candoc= cancer.find(canquery).limit(5) 
for a in candoc:
    pprint(a) 
```

```python
#4) display the first 5 records with diagnosis "Benign"

canquery = { 'diagnosis': 'B' }
candoc= cancer.find(canquery).limit(5)
for b in candoc: 
    pprint(b) 
```

```python
#5) Show the top 10 patient diagnosis with largest area

canquery = cancer.find({},{"area_mean": 1, "diagnosis": 1}).sort("area_mean", -1).limit(10)
for d in canquery: 
    pprint(d) 
```

```python
#6) Display all information for the first patient

findone= cancer.find_one()
pprint(findone) 
```

```python
#7) display records with radius_mean less than 7.7

canquery = { "radius_mean": { "$lt": 7.7 } }
candoc = cancer.find(canquery) 
for f in candoc:
    pprint(f)
```

```python
#8) display records with texture measure greater than 30

canquery = { "texture_mean": { "$gt": 30 } }
candoc = cancer.find(canquery) 
for g in candoc:
    pprint(g)
```

### Aggregation

```python
#9) calculate the sum of concavity measure

agr = [ {'$group': {'_id': 1, 'all': { '$sum': '$concavity_mean' } } } ]
val = list(db.cancer.aggregate(agr)) 
print('The sum of concavity_mean is {}'.format(val[0]['all'])) 
```

```python
#10) calculate the total of fractal dimension of patients with Benign cancer and perimeter is greater than 100

agr = [{ '$match': {'$or': [ { 'diagnosis': "B"  }, { 'perimeter_mean': { "$gt": 100 } }] }}, 
        { '$group': {'_id': 1, 'total': { '$sum': "$fractal_dimension_mean" } }}]
val = list(db.cancer.aggregate(agr)) 

print('The requested value is {} '.format(val[0]['total']))
```

### Update

```python
#11) display the record before updating

canquery = { 'id': 88466802}        
candoc= cancer.find(canquery,{"compactness_mean": 1})
for h in candoc: 
    print(h) 
```

```python
#12) update the record using set

db.cancer.update({"id":88466802},{"$set":{"compactness_mean": 0.0999}})
```

```python
#13) display the record after updating

canquery = { 'id': 88466802} 
candoc= cancer.find(canquery,{"compactness_mean": 1}) 
for h in candoc: 
    print(h) 
```

```python
#14) display the record before updating

canquery = { 'id': 883539}        
candoc= cancer.find(canquery,{"area_mean": 1})
for h in candoc: 
    print(h) 
```

```python
#15) increment the area_mean of specific record by 100

db.cancer.update({"id":883539},{"$inc":{"area_mean":100}})     
```

```python
#16) display the record before updating

canquery = { 'id': 883539}        
candoc= cancer.find(canquery,{"area_mean": 1})
for h in candoc: 
    print(h) 
```
