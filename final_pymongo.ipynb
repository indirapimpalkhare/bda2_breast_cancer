```python
#1) import mongo client

from pymongo import MongoClient
import pandas as pd
```

```python
#2) create an instance of the client

client=MongoClient()
```

```python
#3) establish connection

Client= MongoClient('localhost',27017)
```

```python
#4) create database 

db = client.BDAmini
```

```python
#5) create collection

cancer=db.cancer
```

```python
#6) import the dataset

df=pd.read_csv("data.csv")
```

```python
#1) print total number of patient records

cancer.count_documents({})
```

```python
#2) Display all records
from pprint import pprint

for e in cancer.find():
    pprint(e)
```

```python
#3) display the first 5 records with diagnosis "Malignant"

canquery = { 'diagnosis': 'M' }
candoc= cancer.find(canquery).limit(5) 
for a in candoc:
    pprint(a)
```

```python
#4) display the first 5 records with diagnosis "Benign"

canquery = { 'diagnosis': 'B' }
candoc= cancer.find(canquery).limit(5)
for b in candoc: 
    pprint(b) 
```

```python
#5) Show the top 10 patient diagnosis with largest area

canquery = cancer.find({},{"area_mean": 1, "diagnosis": 1}).sort("area_mean", -1).limit(10)
for d in canquery: 
    pprint(d) 
```

```python
#6) Display all information for the first patient

findone= cancer.find_one()
pprint(findone) 
```

```python
#7) display records with radius_mean less than 7.7

canquery = { "radius_mean": { "$lt": 7.7 } }
candoc = cancer.find(canquery) 
for f in candoc:
    pprint(f)
```

```python
#8) display records with texture measure greater than 30

canquery = { "texture_mean": { "$gt": 30 } }
candoc = cancer.find(canquery) 
for g in candoc:
    pprint(g)
```

```python
#9) calculate the sum of concavity measure

agr = [ {'$group': {'_id': 1, 'all': { '$sum': '$concavity_mean' } } } ]
val = list(db.cancer.aggregate(agr)) 
print('The sum of concavity_mean is {}'.format(val[0]['all'])) 
```

```python
#10) calculate the total of fractal dimension of patients with Benign cancer and perimeter is greater than 100

agr = [{ '$match': {'$or': [ { 'diagnosis': "B"  }, { 'perimeter_mean': { "$gt": 100 } }] }}, 
        { '$group': {'_id': 1, 'total': { '$sum': "$fractal_dimension_mean" } }}]
val = list(db.cancer.aggregate(agr)) 

print('The requested value is {} '.format(val[0]['total']))
```
